```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('C:\\Users\\benec\\Downloads\\archive (8)\\Inc_Exp_Data.csv')
print(df)
```

```python
df=pd.get_dummies(df,columns=['Highest_Qualified_Member'],drop_first=True)
df
```

```python
X=df['Mthly_HH_Expense'].values.reshape(-1,1)
y=df['Mthly_HH_Income']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
plt.scatter(X_train,y_train,color='blue',label='Training data')
plt.scatter(X_test,y_test,color='red',label='Testing data')
plt.plot(X_test,y_pred,color='green',label='Prediction line')
plt.title("Monthly Expenditure")
plt.xlabel('Monthly income')
plt.ylabel('family members')
plt.legend()
plt.show()
```
